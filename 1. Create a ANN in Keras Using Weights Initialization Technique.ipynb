```python
import pandas as pd
df = pd.read_csv('Churn_Modelling.csv')
```

- Our Aim is to predict wheter the customers will exit or not in future

```python
X = df.iloc[:,3:13]
y = df.iloc[:,13]
```

```python
X.head()
```

```python
y.head()
```

```python
# Geography and gender categorical features converting into numbers

geograpy = pd.get_dummies(X['Geography'],drop_first=True)
gender = pd.get_dummies(X['Gender'],drop_first=True)
```

```python
geograpy.head()
```

```python
gender.head()
```

```python
X.head()
```

```python
X = pd.concat([X,geograpy,gender],axis=1)
```

```python
X.head()
```

```python
X= X.drop(['Geography','Gender'],axis=1)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

# Feature Scaling
- Multiplication will happen faster when will scale the values and multiply with weigths 
- Also Global minima can reach to quickly

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
X_train
```

```python
X_test
```

# Create the ANN model

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
#from keras.layers import LeakyReLU, PReLU, ELU
from keras.layers import Dropout
```

## Sequential : Responsible for creating the ANN,CNN,RNN models
## Dense: To create the Hidden layers
## LeakyRelu, Prelu, ELU are the activations function
## Dropout is a regularization parameters it will be used when ANN is deep

```python
classifier =Sequential()  # Empty Neural Network
```

```python
classifier.add(Dense(units = 10,kernel_initializer='he_normal',activation='relu',input_dim=11))
classifier.add(Dropout(0.3)) # after appling in this NN accuracy drop to 3-4 %, so use it only when u have Deep NN
# here creating the first HL
# output dim is 6 which means 6 Hidden neurons in first layer
# initialization parameter how your weights needs to be initialized
# whenever using relu AF he uniform and he normal will work much better
# relu prevents Vanishing gradient problem as compare to sigmoid
# All Hidden layers should have relu or Leakyrelu
# 
# input_dim is 11 
```

```python
# second Hidden Layer
classifier.add(Dense(units = 20,kernel_initializer='he_normal',activation='relu'))
classifier.add(Dropout(0.4))
```

```python
# Third Hidden Layer
classifier.add(Dense(units = 15,kernel_initializer='he_normal',activation='relu'))
classifier.add(Dropout(0.2))
```

```python
# Final Layer 
classifier.add(Dense(units=1, kernel_initializer='glorot_uniform', activation='sigmoid'))
```

```python
classifier.summary()
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model_history = classifier.fit(X_train,y_train,validation_split=0.33, batch_size = 10, epochs =100)
```

```python
# Prediction on test Data
y_pred = classifier.predict(X_test)
y_pred = (y_pred>0.5)
```

```python
y_pred
```

```python
# Confusion metrics

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
# Accuracy
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred)
```
